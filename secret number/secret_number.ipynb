```python
import random
```

```python
secret_number = random.randint(1, 100)
```

```python
tentativas = 0
```

```python
while True:
    palpite = int(input("Insira sua tentativa: "))
    tentativas += 1
    if palpite > secret_number:
        print ("O numero digitado e maior do que o numero correto")
    elif palpite < secret_number:
        print("O numero digitado e menor do que o numero correto")
    else:
        print("Numeros iguais!")
        print(f"Voce precisou de {tentativas} tentativas")
        break
```
