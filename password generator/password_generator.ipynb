```python
import random
import string
```

```python
tamanho = int(input("Insira o tamanho da senha: "))
```

```python
letra = (input("Letras maiusculas? (s/n"))
numero = (input("Numeros? (s/n)"))
simbolo = (input("Simbolos? (s/n)"))
```

```python
caractere = string.ascii_lowercase
```

```python
if letra == "s":
    caractere += string.ascii_lowercase
if numero == "s":
    caractere += string.digits
if simbolo == "s":
    caractere += string.punctuation
```

```python
senha = ""
```

```python
for i in range (tamanho):
    senha += random.choice(caractere)
```

```python
print(senha)
```
