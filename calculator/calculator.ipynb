```python
numero1 = float(input("Insira um numero: "))
numero2 = float(input("Insira outro numero: "))
operacao = input("Selecione a operacao: +, -, *, /")
```

```python
if operacao == "+":
    resultado = numero1 + numero2
    print(resultado)

elif operacao == "-":
    resultado = numero1 - numero2
    print(resultado)

elif operacao == "*":
    resultado = numero1 * numero2
    print(resultado)

elif operacao == "/":
    if numero2 != 0:
        resultado = numero1 / numero2
        print(resultado)
    else:
        print("Erro.")

else:
    print("Operacao invalida.")
```
